# Useful functions 

```python
def outlier_2nan(df, factor=1.5, action='drop'):
    '''
    outlier_2nan Substituição de outliers.

    Args:
        factor (float): Fator de correção.
        action (text object): drop ou mean. Regra de substituição dos outliers encontrados.
    '''
    for c in df.columns:
        q3, q1 = np.percentile(df[c], [75, 25])
        iqr = q3 - q1
        lowpass = q1 - (iqr*factor)
        highpass = q3 + (iqr*factor)
        df.loc[(df[c] < lowpass) | (df[c] > highpass), c] = np.nan

    if action == 'drop':
        df.dropna(axis=0, inplace=True)
    if action == 'mean':
        df.fillna(df.mean(), axis=0, inplace=True)
```

```python
def corr_map(data, method):
    '''
    corr_map Plota correlação de colunas em pares, excluindo valores NA/nulos.

    Args:
        data (pandas DataFrame): Tabela contendo as observações.
        method (text object): Method of correlation: pearson, kendall ou spearman.
    '''
    corr=data.corr(method=method)
    mask = np.zeros_like(corr)
    mask[np.triu_indices_from(mask)] = True
    plt.figure(figsize=(10,10))
    sns.heatmap(corr,cmap="coolwarm", linewidths=0.5, annot=True, annot_kws={"size":8},mask=mask,square=True,cbar=False)
    plt.show()
```

```python
def time_features(df, w, cols):
    '''
    time_features Criação de novas variáveis estatísticas a partir de uma janela deslizante: 
                  Mínimo, Máximo, Média, Mediana, Variância, Desvio-padrão, Assimetria, Curtose e Autocorrelação lag 1.
    
    Args:
        df (pandas DataFrame): Tabela contendo observações
        w (int): Tamanho da janela em movimento. Este é o número de observações utilizadas para o cálculo da estatística.
        cols (text list): Colunas em que a criação de variáveis será empregada.
    '''
    for c in cols:
        data = df[c]
        aux = pd.DataFrame({'min': data.rolling(window=w).min(),
                            'max': data.rolling(window=w).max(),
                            'mean': data.rolling(window=w).mean(),
                            'median': data.rolling(window=w).median(),
                            'var': data.rolling(window=w).var(),
                            'std': data.rolling(window=w).std(),
                            'skew': data.rolling(window=w).skew(),
                            'kurt': data.rolling(window=w).kurt(),
                            'corr1': data.rolling(window=w).apply(lambda x: x.autocorr(lag=1), raw=False)})
        df['{}_min_{}'.format(c, str(w))] = aux['min'].fillna(data)
        df['{}_max_{}'.format(c, str(w))] = aux['max'].fillna(data)
        df['{}_mean_{}'.format(c, str(w))] = aux['mean'].fillna(data)
        df['{}_median_{}'.format(c, str(w))] = aux['median'].fillna(data)
        df['{}_var_{}'.format(c, str(w))] = aux['var'].fillna(data)
        df['{}_std_{}'.format(c, str(w))] = aux['std'].fillna(data)
        df['{}_skew_{}'.format(c, str(w))] = aux['skew'].fillna(data)
        df['{}_kurt_{}'.format(c, str(w))] = aux['kurt'].fillna(data)
        df['{}_corr1_{}'.format(c, str(w))] = aux['corr1'].fillna(data)
```

```python
import numpy as np
from scipy.fftpack import fft

def FFT(time, data):
    '''
    FFT Cálculo da Transformada discreta de Fourier 1-D.
    
    Args:
        time (numpy array): Array contendo as observações temporais da entrada.
        data (numpy array): Array de entrada, pode ser complexo.
    
    Returns:
        numpy array: Array contendo os valores de frequência da amostragem.
        numpy array: Array contendo os coeficientes da transformada.
    '''
    dt = np.mean(time[1:] - time[:-1])
    N = len(data)
    N2 = 2 ** (int(np.log2(N)) + 1) # round up to next highest power of 2
    freq = np.linspace(0.0, 1.0/(2.0*dt), N2//2)
    coef_ = fft(data)
    coef = 2.0/N2 * np.abs(coef_[0:N2//2])
    return freq, coef
```

```python
import numpy as np
import pywt

# https://la.mathworks.com/help/wavelet/gs/interpreting-continuous-wavelet-coefficients.html
# https://pywavelets.readthedocs.io/en/latest/ref/cwt.html
def CWT(time, data, scales, wavelet):
    '''
    CWT Transformada wavelet contínua unidimensional.
    
    Args:
        time (numpy array): Time information of signal.
        data (numpy array): Input signal.
        scales (numpy array): The wavelet scales to use. An array of float or int >= 1 with increasing values
                              One can use to determine what physical frequency. 
                              Here, is in hertz when the is given in seconds.
                              f = scale2frequency(wavelet, scale)/sampling_periodffsampling_period
        wavelet (Wavelet object or name): Wavelet to use.

    Returns:
    numpy array: Continuous wavelet transform of the input signal for the given scales and wavelet. 
                 The first axis of corresponds to the scales. The remaining axes match the shape of coefsdata
    numpy array: If the unit of sampling period are seconds and given, than frequencies are in hertz. 
                 Otherwise, a sampling period of 1 is assumed.
    '''
    dt = np.mean(time[1:] - time[:-1])
    coef, freq = pywt.cwt(data, scales, wavelet, dt)
    return freq, coef
```

```python
def plot_data(ax, time, data, legend=False, title=None):
    '''
    plot_data Plot do Sinal
    
    Args:
        ax (matplotlib.axes)
        time (numpy aray): Informação temporal do sinal de entrada (eixo-x).
        data (numpy aray): Sinal de entrada.
        legend (bool): Indicar se a legenda "data" deve ser apresentada. Default é False.
        title (text object): Título do gráfico. Default é None
    '''
    ax.plot(time, data, label='data')
    ax.set_xlim([time.min(), time.max()])
    if legend:
        ax.legend(loc='upper right')
    if title:
        ax.set_title(title)
```

```python
def plot_fft_plus_power(ax, time, data, plot_direction='horizontal', 
                        yscale=None, avline=None, legend=False, 
                        xlabel=False, ylabel=False, title=False):
    '''
    plot_fft_plus_power Plot FFT e o Espectro de Potência (np.var(data)*abs(coef)**2).

    Args:
        ax (matplotlib.axes)
        time (numpy aray): Informação temporal do sinal de entrada (eixo-x).
        data (numpy aray): Sinal de entrada.
        plot_direction (): vertical é utilizado para o gráfico composto da transformada wavelet e fourier. Default é horizontal.
        yscale (text object): O axis scale a ser aplicado: linear, log, symlog, logit, ... Default é None.
        avline (float list): Lista contendo valores de linhas verticais a serem adicionadas ao plot. 
                             Útil para observar filtros aplicados aos dados. Default é None.
        legend (bool): Show "Fourier Transform" e "FFT Power Spectrum". Default é False.
        xlabel (bool): Show "Frequency". Default é False.
        ylabel (bool): Show "Spectrum". Default é False.
        title (bool): Show "Fourier Transform". Default é False.
    '''
    freq, coef = FFT(time, data)
    
    variance = np.std(data)**2
    power = variance*abs(coef)**2

    if plot_direction == 'horizontal':
        if yscale:
            valid = (coef > 0) & (power > 0)
            freq, coef, power = freq[valid], coef[valid], power[valid]
            ax.set_yscale(yscale);
        
        ax.plot(freq, coef, 'r-', label='Fourier Transform')
        ax.plot(freq, power, 'k--', linewidth=1, label='FFT Power Spectrum')
        ax.set_xlim(freq.min(), freq.max())
        ax.set_ylim(min(coef.min(), power.min()), max(coef.max(), power.max()))
        
        if avline:
            for a in avline:
                ax.axvline(x=a, color='k')
    elif plot_direction == 'vertical':
        if yscale:
            valid = freq > 0
            freq, coef, power = freq[valid], coef[valid], power[valid]
            ax.set_yscale(yscale);
        
        ax.plot(coef, freq, 'r-', label='Fourier Transform')
        ax.plot(power, freq, 'k--', linewidth=1, label='FFT Power Spectrum')
        ax.set_xlim(min(coef.min(), power.min()), max(coef.max(), power.max()))
        ax.set_ylim(freq.min(), freq.max())
        #ax.invert_yaxis()
    
    if title:
        ax.set_title("Fourier Transform")
    if xlabel:
        ax.set_xlabel("Frequency")
    if ylabel:
        ax.set_ylabel("Spectrum")
    if legend:
        ax.legend()
```

```python
def plot_wavelet(ax, time, data, scales, wavelet,
                 spectrum='amp', yaxis='period', cmap='jet',
                 cbar='vertical', clim=None, cbarlabel=False, cscale='linear',
                 yscale=None, xlabel=False, ylabel=False, title=False):
    '''
    plot_wavelet Build and displays the 2D spectrum for Continuous Wavelet Transform.
                 Build the CWT and displays the scaleogram.

    Args:
        ax (matplotlib.axes)
        time (numpy array): Time information of signal.
        data (numpy array): Input signal.
        scales (numpy array): The wavelet scales to use. An array of float or int >= 1 with increasing values.
        wavelet (Wavelet object or name): Wavelet to use.

        spectrum (text object): selects the type of processing for the CWT data.
                                - 'amp' displays abs(CWT) (L1 norm)
                                - 'real' / 'imag' for real or imaginary part only
                                - 'power' for abs(CWT)**2
        yaxis (text object): selects the Y axis units.
                             - 'period' Convert scales to human readable period values. 
                                        The period units is the same as the time input parameter.
                                        If time is not provided, periods are in number of samples units.
                             - 'frequency' : Converts scales to frequency.
                                             The frequency unit is depending on the time argument value.
                                             If time is not provided, the frequency represents the number of oscillations per sample.
                                             In this mode yscale is set to 'log' by default (if not provided).
                             - 'scale': display the wavelet scales parameter.                        
        cmap (matplotlib color map name (or instance)): matplotlib color map name (or instance) for the spectrum.
                                                       To view available maps type: matplotlib.pyplot.colormaps().
        cbar (text object): 'vertical' | 'horizontal': selects the color bar location.
        clim (min, max) : sets the color range limits for spectrum values (in other words: z axis range).
        cscale (text object) 'linear' | 'log':  selects scaling for CWT spectrum.
                                                 This parameter impacts the color bar ticks values and spectrum colors.
        yscale (text object) 'linear' | 'log':  selects scaling for the Y axis.
'''
    scales_freq, coefs = CWT(time, data, scales, wavelet)
    
    # adjust y axis ticks
    scales_period = 1./scales_freq
    dt = np.mean(time[1:] - time[:-1])
    xmesh =  np.concatenate([time, [time[-1]+dt]])
    if yaxis == 'period':
        ymesh = np.concatenate([scales_period, [scales_period[-1]+dt]])
        if ylabel:
            ax.set_ylabel("Period")
    elif yaxis == 'frequency':
        df = scales_freq[-1]/scales_freq[-2]
        ymesh = np.concatenate([scales_freq, [scales_freq[-1]*df]])
        # set a useful yscale default: the scale freqs appears evenly in logscale
        #yscale = 'log' if yscale is None else yscale
        if ylabel:
            ax.set_ylabel("Frequency")
        #ax.invert_yaxis()
    elif yaxis == 'scale':
        ds = scales[-1]-scales[-2]
        ymesh = np.concatenate([scales, [scales[-1] + ds]])
        if ylabel:
            ax.set_ylabel("Scale")
    else:
        raise Exception("yaxis undefined!")

    ax.set_xlim(time.min(), time.max())
    ax.set_ylim(ymesh.min(), ymesh.max())
    
    if yscale:
        ax.set_yscale(yscale)

    # choose the correct spectrum display function and name
    if spectrum == 'amp':
        values = np.abs(coefs)
        sp_title = "Amplitude"
        cbarlabel= "abs(CWT)" if cbarlabel is None else cbarlabel
    elif spectrum == 'real':
        values = np.real(coefs)
        sp_title = "Real"
        cbarlabel= "real(CWT)" if cbarlabel is None else cbarlabel
    elif spectrum == 'imag':
        values = np.imag(coefs)
        sp_title = "Imaginary"
        cbarlabel= "imaginary(CWT)" if cbarlabel is None else cbarlabel
    elif spectrum == 'power':
        sp_title = "Power"
        cbarlabel= "abs(CWT)$^2$" if cbarlabel is None else cbarlabel
        values = np.power(np.abs(coefs),2)
    else:
        raise Exception("spectrum undefined!")

    # labels and titles
    if title:
        ax.set_title("Continuous Wavelet Transform "+sp_title+" Spectrum")
    if xlabel:
        ax.set_xlabel("Time/spatial domain")

    if cscale == 'log':
        isvalid = (values > 0)
        cnorm = LogNorm(values[isvalid].min(), values[isvalid].max())
    else:
        cnorm = None

    #qmesh = ax.pcolormesh(xmesh, ymesh, values, cmap=cmap, norm=cnorm)
    qmesh = ax.contourf(xmesh[:-1], ymesh[:-1], values, cmap=cmap, norm=cnorm)

    if cbar:
        if cbar == 'vertical':
            cbarkw = {'aspect':30, 'pad':0.03, 'fraction':0.05}
        elif cbar == 'horizontal':
            cbarkw = {'aspect':40, 'pad':0.12, 'fraction':0.05}
        colorbar = plt.colorbar(qmesh, orientation=cbar, ax=ax, **cbarkw)
        if cbarlabel:
            colorbar.set_label(cbarlabel)
```

```python
import matplotlib.ticker
import matplotlib.gridspec as gridspec
import matplotlib.patches as patches

def plot_data_fft_wavelet(time, data, scales, wavelet, figsize=(12,12)):
    # Plot signal FFT and Wavelet transform in same figure.

    fig, ax = plt.subplots(figsize=figsize)
    ax.set_axis_off()
    spec = gridspec.GridSpec(ncols=6, nrows=6)
    top_ax = fig.add_subplot(spec[0, 0:5])
    bottom_left_ax = fig.add_subplot(spec[1:, 0:5])
    bottom_right_ax = fig.add_subplot(spec[1:, 5])

    yscale = 'log'
    plot_data(top_ax, time, data, legend=False)
    plot_fft_plus_power(bottom_right_ax, time, data, plot_direction='vertical', 
                        yscale=yscale, avline=None, legend=False, 
                        xlabel=False, ylabel=False, title=False)
    plot_wavelet(bottom_left_ax, time, data, scales, wavelet,
                 spectrum='amp', yaxis='frequency', cbar=False, yscale=yscale, xlabel=False, ylabel=False, title=False)
    
    # align frequency values
    lw = bottom_left_ax.get_ylim()
    bottom_right_ax.set_ylim(lw)
    bottom_right_ax.axes.get_yaxis().set_visible(False)
    
    # align period values
    top_ax.axes.get_xaxis().set_visible(False)

    # final adjust
    plt.tight_layout()
```

```python
import numpy as np
import pywt

def lowpassfilter_wavedec(data, thresh, wavelet='db6'):
    # Low pass filter using wavelet
    # https://pywavelets.readthedocs.io/en/latest/ref/dwt-discrete-wavelet-transform.html#multilevel-decomposition-using-wavedec 
    # https://stackoverflow.com/questions/54619107/how-to-combine-wavelet-transform-and-frequency-filtering

    thresh = thresh*np.nanmax(data)
    coeff = pywt.wavedec(data, wavelet, mode='symmetric')
    coeff[1:] = (pywt.threshold(i, value=thresh, mode='soft') for i in coeff[1:])
    reconstructed_data = pywt.waverec(coeff, wavelet, mode='symmetric')
    return reconstructed_data
```

```python
import scipy.signal as signal
from math import pi

# https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.butter.html
def butterworth(data, w):
    '''
    butterworth Butterworth low pass filter.
    
    Args:
        data (numpy array): Input signal.
        w (float): Cut-off frequency of the butterworth filter, normalized by frequency.
                   For digital filters, Wn are in the same units as fs. 
                   By default, fs is 2 half-cycles/sample, so these are normalized from 0 to 1, where 1 is the Nyquist frequency. (Wn is thus in half-cycles/sample.)
                   fc = pi/5  # Cut-off frequency of the butterworth filter
                   w = fc/2 # Normalize the frequency
    '''
    b, a = signal.butter(4, w, 'low')
    filtered_data_butterworth = signal.filtfilt(b, a, data)
    return filtered_data_butterworth
```

# Importing dataset

The objective of the work is to map instrumented railway vehicle (IRV) data to track irregularities in order to understand the underlying relationship between these parameters. The dataset is generated through simulation of a virtual IRV that measures the same variables as the real IRV. The simulation was done in the software SIMPACK and the track irregularities are generated with a power spectral density (PSD) function. 


The final dataset can be summarized in two categories:

1. IRV dataset (input variables):
  -  Carbody acceleration - X component - $\left\{\frac{m}{s^2}\right\}$
  -  Carbody acceleration - Y component - $\left\{\frac{m}{s^2}\right\}$
  -  Carbody acceleration - Z component - $\left\{\frac{m}{s^2}\right\}$
  -  Leading bogie vertical acceleration - left sideframe - $\left\{\frac{m}{s^2}\right\}$
  -  Leading bogie vertical acceleration - right sideframe - $\left\{\frac{m}{s^2}\right\}$
  -  Trailing bogie vertical acceleration - left sideframe - $\left\{\frac{m}{s^2}\right\}$
  -  Trailing bogie vertical acceleration - right sideframe - $\left\{\frac{m}{s^2}\right\}$
  -  Leading bogie secondary suspension displacement (deslocamento da suspensão secundária - Disp) - left sideframe - $\left\{m\right\}$
  -  Leading bogie secondary suspension displacement - right sideframe - $\left\{m\right\}$
  -  Trailing bogie secondary suspension displacement - left sideframe - $\left\{m\right\}$
  -  Trailing bogie secondary suspension displacement - left sideframe - $\left\{m\right\}$
  -  Time (used for FFT) - $\left\{s\right\}$
  -  Distance (used for alignment) - $\left\{m\right\}$

2. Irregularity dataset (output variables):
  -  Vertical track irregularity of the left rail - $\left\{m\right\}$
  -  Vertical track irregularity of the right rail - $\left\{m\right\}$
  -  Lateral track irregularity of the left rail - $\left\{m\right\}$
  -  Lateral track irregularity of the right rail - $\left\{m\right\}$
  -  Distance (used for alignment) - $\left\{m\right\}$

# Feature engineering

## Domain knowledge variables

image.png

```python
z_accel_columns = ['Accel_uni_Z_lead_bogie_right', 'Accel_uni_Z_lead_bogie_left', 'Accel_uni_Z_trail_bogie_right', 'Accel_uni_Z_trail_bogie_left']
susp_columns = ['Disp_lead_bogie_left', 'Disp_lead_bogie_right', 'Disp_trail_bogie_left', 'Disp_trail_bogie_right']
```

```python
logic_vector = df[z_accel_columns].values.max(1) > -1*df[z_accel_columns].values.min(1)

df['Z_accel_max'] = df[z_accel_columns].values.min(1) # Initializing the vector
df['Z_accel_max'][logic_vector] = df[z_accel_columns][logic_vector].values.max(1) # Overwriting the vector
```

```python
susp_col_left = ['Disp_lead_bogie_right','Disp_trail_bogie_right']
susp_col_right = ['Disp_lead_bogie_left','Disp_trail_bogie_left']

logic_vector_left= df[susp_col_left].values.max(1) > -1*df[susp_col_left].values.min(1)
df['Susp_travel_left'] = df[susp_col_left].values.min(1) 
df['Susp_travel_left'][logic_vector_left] = df[susp_col_left][logic_vector_left].values.max(1)


logic_vector_right= df[susp_col_right].values.max(1) > -1*df[susp_col_right].values.min(1)
df['Susp_travel_right'] = df[susp_col_right].values.min(1) 
df['Susp_travel_right'][logic_vector_right] = df[susp_col_right][logic_vector_right].values.max(1)


col = ['Susp_travel_right','Susp_travel_left']

logic_vector= df[col].values.max(1) > -1*df[col].values.min(1)
df['Susp_travel'] = df[col].values.min(1) 
df['Susp_travel'][logic_vector] = df[col][logic_vector].values.max(1)
```

```python
df['Front_Bounce'] = (df['Disp_lead_bogie_left']+df['Disp_lead_bogie_right'])/2
df['Back_Bounce'] = (df['Disp_trail_bogie_left']+df['Disp_trail_bogie_right'])/2

col = ['Front_Bounce','Back_Bounce']

logic_vector = df[col].values.max(1) > -1*df[col].values.min(1)
df['Bounce'] = df[col].values.min(1)
df['Bounce'][logic_vector] = df[col][logic_vector].values.max(1)
```

```python
df['Front_Roll'] = (df['Disp_lead_bogie_left']-df['Disp_lead_bogie_right'])
df['Back_Roll'] = (df['Disp_trail_bogie_left']-df['Disp_trail_bogie_right'])

col = ['Front_Roll','Back_Roll']

logic_vector = df[col].values.max(1) > -1*df[col].values.min(1)
df['Roll'] = df[col].values.min(1)
df['Roll'][logic_vector] = df[col][logic_vector].values.max(1)
```
